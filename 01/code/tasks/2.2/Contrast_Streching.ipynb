```python
import numpy as np
import matplotlib.pyplot as plt

def contrast_stretching(image):
    # Define the minimum and maximum desired output intensity values (e.g., 0 and 255)
    min_output = 0     #alpha
    max_output = 255   #beta
    
    # Separate the RGB channels
    red_channel, green_channel, blue_channel = image[:,:,0], image[:,:,1], image[:,:,2]

    fig 

    # Apply contrast stretching independently to each channel
    red_stretched = (red_channel - np.min(red_channel)) * (max_output / (np.max(red_channel) - np.min(red_channel))) + min_output
    green_stretched = (green_channel - np.min(green_channel)) * (max_output / (np.max(green_channel) - np.min(green_channel))) + min_output
    blue_stretched = (blue_channel - np.min(blue_channel)) * (max_output / (np.max(blue_channel) - np.min(blue_channel))) + min_output
    
    # Merge the stretched channels back into an RGB image
    stretched_image = np.dstack((red_stretched, green_stretched, blue_stretched)).astype(np.uint8)

    return stretched_image

# Load an RGB image (replace 'your_image.jpg' with the actual image file path)
image = plt.imread('CV_images/data/example1.png')

# Apply contrast stretching
stretched_image = contrast_stretching(image)

# Display the original and stretched images
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(stretched_image)
plt.title('Contrast Stretched Image')

plt.show()
    
```
