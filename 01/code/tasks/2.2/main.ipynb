```python
import imageio
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def read_img(img_path):
    return imageio.imread(img_path)
```

```python
img_path = "../../../images/used_images/light_dark/"
img_name = "DSC_0021.JPG"

img_rgb = read_img(img_path+img_name)
img_hsi = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2HLS)
```

```python
fig, axs = plt.subplots(1, 2)

axs[0].imshow(img_rgb)
axs[1].imshow(img_hsi)
fig.suptitle('RGB vs HLS')
fig.set_size_inches(18.5, 5.5)
plt.show()
```
